```python
import pycaret
```

```python
from pycaret.datasets import get_data
```

```python
df = get_data("result/greedy")
```

```python
df_train = df.sample(frac=0.9, random_state=984)
df_test = df.drop(df_train.index)
df.shape
df_train.head()
```

```python
df_train.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)
```

```python
from pycaret.classification import *
s = setup(df_train, target = 'Classes')
```

```python
all_models = ['lr','knn','nb','dt','svm','rbfsvm','gpc','ridge','rf','ada','gbc','lda','et','dummy']
prefer_models = ['dt','rbfsvm','gbc','rf','et','knn']
# best = compare_models(include=prefer_models)
# best = compare_models(include=all_models)
```

```python
for idx, item in enumerate(prefer_models):
    model = create_model(item)
    evaluate_model(model)
    save_model(model, f"model/model_{item}")
```
